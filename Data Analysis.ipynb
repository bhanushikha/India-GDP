```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
econ_data = history.response[0]  # Assuming there's only one element in the list
print(dir(econ_data))

# Create DataFrame
df = pd.DataFrame(econ_data.series)
    
# Print or use the DataFrame as needed
print(df)
```

```python
Finaldata = df[['date', 'displayValue']]
```

```python
print(Finaldata)
```

```python
Finaldata.loc[:, 'date'] = pd.to_datetime(Finaldata['date'])
```

```python
# Filter the DataFrame for the last 10 years
cutoff_date = Finaldata['date'].max() - pd.DateOffset(years=10)
filtered_df = Finaldata[Finaldata['date'] >= cutoff_date]

# Sort the filtered DataFrame by date
filtered_df = filtered_df.sort_values(by='date')

# Plot the filtered time series
plt.plot(filtered_df['date'], filtered_df['displayValue'], marker='o')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series of Last 10 Years')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add grid lines
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
